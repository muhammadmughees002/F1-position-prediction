```python
import pandas as pd 
```

```python
# Load your dataset
df = pd.read_csv("results.csv")
```

```python
df
```

```python
# Drop rows with missing target
df = df[df['positionOrder'].notna()]
```

```python
# Define target and features
target = 'positionOrder'
```

```python
# Select features (excluding target and IDs you don't want to leak info)
features = ['driverId', 'constructorId', 'grid', 'laps', 'milliseconds',
            'fastestLap', 'rank', 'fastestLapSpeed', 'statusId']
```

```python
X = df[features]
y = df[target]
```

```python
#encoding information
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Make a copy to avoid SettingWithCopyWarning
X = X.copy()
```

```python
# Encode driverId, constructorId, statusId
label_encoders = {}
for col in ['driverId', 'constructorId', 'statusId']:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col].astype(str))
    label_encoders[col] = le
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# training random forest 
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)
```
